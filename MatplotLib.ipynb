```python
import matplotlib.pyplot as plt
```

# scatterplot and plot

```python
x=[1,2,3]
y=[2,1,5]
plt.scatter(x,y) # points out the co-ordinates
plt.plot(x,y)    # Joins the co-ordinates with lines
plt.xlabel('cost')
plt.ylabel('Apples')
plt.title('Cost of apples')
plt.legend()
plt.show()
```

```python
x1=[1,2,3]
y1=[2,1,5]
plt.scatter(x1,y1) # points out the co-ordinates
plt.plot(x1,y1,label='Label1')    # Joins the co-ordinates with lines
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title Of The Graph')
x2=[10,-2,3]
y2=[21,-1,5]
#plt.scatter(x2,y2) # points out the co-ordinates
plt.plot(x2,y2,label='Label2') 
plt.legend()
plt.show()
plt.close()
```

```python
print(plt.style.available)
x=[1,2,3,4,8]
y=[2,1,5,-1,-3]
plt.plot(x,y,linestyle='dashed',linewidth=5,marker='*',markerfacecolor='red',markersize=10)
plt.style.use('fivethirtyeight')
plt.xlim(-1,10)
plt.ylim(-5,10)
plt.show()
```
