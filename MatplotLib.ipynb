```python
import matplotlib.pyplot as plt
```

# scatterplot and plot

```python
x=[1,2,3]
y=[2,1,5]
plt.scatter(x,y) # points out the co-ordinates
plt.plot(x,y)    # Joins the co-ordinates with lines
plt.xlabel('cost')
plt.ylabel('Apples')
plt.title('Cost of apples')
plt.legend()
plt.show()
```

```python
x1=[1,2,3]
y1=[2,1,5]
plt.scatter(x1,y1) # points out the co-ordinates
plt.plot(x1,y1,label='Label1')    # Joins the co-ordinates with lines
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title Of The Graph')
x2=[10,-2,3]
y2=[21,-1,5]
#plt.scatter(x2,y2) # points out the co-ordinates
plt.plot(x2,y2,label='Label2') 
plt.legend()
plt.show()
plt.close()
```

```python
print(plt.style.available)
x=[1,2,3,4,8]
y=[2,1,5,-1,-3]
plt.plot(x,y,linestyle='dashed',linewidth=5,marker='*',markerfacecolor='red',markersize=10)
plt.style.use('fivethirtyeight')
plt.xlim(-1,10)
plt.ylim(-5,10)
plt.show()
```

# Bar Chart

```python
left=[5,4,3,2,1]
height=[13,20,37,40,60]
labelX=['One','Two','Three','Four','Five']
plt.bar(left,height,tick_label=labelX,color=['red','blue'],width=0.5)
```

# Histogram

```python
ages=[2,3,5,1,6,8,0,3,6,5,87,64,28,59,30,34,75,24,64,96,75,48,39,10,56,29,4,29,18,16,30,49,67]
range=(0,100)
bins=10
plt.hist(ages,bins,range,color=['blue'],rwidth=0.5,histtype='bar')
```

# Pie Chart

```python
acts=['eat','sleep','work','play']
hrs=[2,10,5,8]
col=['red','blue','green','pink']
plt.pie(hrs,labels=acts,colors=col,radius=2,shadow=True,explode=(0.1,0,0.5,0),autopct='%1.1f%%')
```

# Plot Sin curve

```python
import numpy as np
```

```python
x=np.arange(0,2*(np.pi),0.01)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,label='sin')
plt.plot(x,y2,label='cos')
plt.legend()
```

# Sub plots

```python
x=np.arange(-10,10,0.1)
y1=x
y2=x**2
y3=x**3
y4=np.sqrt(x)
plt.subplot(2,2,1)
plt.plot(x,y1,label='y_1=x',color='red')
plt.subplot(2,2,2)
plt.plot(x,y2,label='y_2=x^2',color='green')
plt.subplot(2,2,3)
plt.plot(x,y3,label='y_3=x^3',color='blue')
plt.subplot(2,2,4)
plt.plot(x,y4,label='y_4=x^0.5',color='red')
plt.show()
```
